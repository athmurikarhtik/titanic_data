```python
sex=pd.get_dummies(titanic_data['Sex'],drop_first=True)
```

```python
embark=pd.get_dummies(titanic_data['Embarked'],drop_first=True)
```

```python
pclass=pd.get_dummies(titanic_data['Pclass'],drop_first=True)
```

```python
titanic_data=pd.concat([titanic_data,sex,embark,pclass],axis=1)
```

```python
titanic_data.drop(['Sex','Embarked','Pclass','PassengerId','Ticket'],axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.drop(['Name'],axis=1,inplace=True)
```

## Train Data

```python
x=titanic_data.drop(["Survived"],axis=1)
```

```python
y=titanic_data["Survived"]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python

X_train,X_test,y_train,y_test = train_test_split(x, y, stratify=y, test_size= 0.15,random_state=1)
```

```python
# Logistic Regression
modelLogistic = LogisticRegression()
modelLogistic.fit(X_train,y_train)
predLogistic = modelLogistic.predict(X_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test,predLogistic)
cnf_matrix
print("Accuracy Logistic Regression:",metrics.accuracy_score(y_test, predLogistic))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# DecisionTreecClassifier
modelDecision = DecisionTreeClassifier()
modelDecision.fit(X_train,y_train)
pred= modelDecision.predict(X_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test,pred)
cnf_matrix
print("Accuracy Decision tree:",metrics.accuracy_score(y_test,pred))
```

```python
# RandomForestClassifier 
modelRF = RandomForestClassifier()
modelRF.fit(X_train,y_train)
pred= modelRF.predict(X_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test,pred)
cnf_matrix
print("Accuracy Decision tree:",metrics.accuracy_score(y_test,pred))
```

```python
# RandomForestClassifier 
modelknn = KNeighborsClassifier()
modelknn.fit(X_train,y_train)
pred= modelknn.predict(X_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test,pred)
cnf_matrix
print("Accuracy Decision tree:",metrics.accuracy_score(y_test,pred))
```

```python
from sklearn.svm import SVC
```
